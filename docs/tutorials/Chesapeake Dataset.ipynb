```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import sys
sys.path.append("../..")
import os

import numpy as np
import matplotlib.pyplot as plt

from torch.utils.data import DataLoader

from torchgeo.datasets import ChesapeakeMD, ChesapeakeDE, ChesapeakeWV, ChesapeakeNY, ChesapeakePA, ChesapeakeDC, ChesapeakeVA, Chesapeake13, Chesapeake7
from torchgeo.samplers import RandomGeoSampler
```

```python
ROOT_DIR = os.path.expanduser("~/mount/data/")
```

## Visualization example

```python
dataset = ChesapeakeDC(ROOT_DIR, download=True)
```

```python
sampler = RandomGeoSampler(
    roi=dataset.bounds,
    size=256,
    length=5
)
```

```python
for bounds in sampler:
    sample = dataset[bounds]
    dataset.plot(sample["masks"])
```

## DataLoader example

```python
sampler = RandomGeoSampler(
    roi=dataset.bounds,
    size=256,
    length=48
)
```

```python
dataloader = DataLoader(dataset, sampler=sampler, batch_size=32)
```

```python
for batch_idx, batch in enumerate(dataloader):
    print(batch["masks"].shape)
```

## Visualization example

```python
def show_image(sample):
    im = sample["image"]
    im = np.rollaxis(im.numpy(),0,3).copy() # convert to HxWxC format
    im = im[:,:,:3]
        
    plt.figure(figsize=(4,4))
    plt.imshow(im)
    plt.axis("off")
    plt.show()
    plt.close()
```

```python
ds = NAIP(
    "tmp/",
    crs="EPSG:26918"
)
```

```python
sampler = RandomGeoSampler(
    roi=ds.bounds,
    size=256,
    length=5
)
```

```python
for bounds in sampler:
    try:
        sample = ds[bounds]
        show_image(sample)
        ds.plot(sample["image"])
        print("--")
    except:
        pass
```

## Visualization example

```python
dataset = So2Sat(
    ROOT_DIR,
    split="train",
    checksum=False
)
```

```python
def show_image(sample, scale_factor=4):
    im = sample["image"]
    im = np.rollaxis(im.numpy(),0,3).copy() # convert to HxWxC format
    im = im[:,:,[10,9,8]] # select the RGB bands
    im = np.clip(im*scale_factor, 0, 1) # scale up the values for better pictures and make sure they are in [0,1]
    
    plt.figure()
    plt.imshow(im)
    plt.title(idx)
    plt.axis("off")
    plt.show()
    plt.close()
```

```python
N = len(dataset)
for idx in np.random.randint(0, N, size=(10)):
    show_image(dataset[idx])
```
